```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
state = pd.read_csv("State-wallets-with-pieces.csv", names=["wallets","pieces"])
random = pd.read_csv("RandomSelector-wallets-with-pieces.csv", names=["wallets","pieces"])
space = pd.read_csv("SpaceSelector-wallets-with-pieces.csv", names=["wallets","pieces"])
```

```python

piecesByWallet = pd.merge(state.groupby(["pieces"]).sum().rename(columns={"wallets":"state"}),random.groupby(["pieces"]).sum().rename(columns={"wallets":"random"}),how="left",on="pieces")
piecesByWallet = pd.merge(piecesByWallet,space.groupby(["pieces"]).sum().rename(columns={"wallets":"space"}),on="pieces",how="left")
```

```python
s = piecesByWallet.reset_index()
```

```python
plt.figure(figsize=(20,15))


# fig, ax = plt.subplots()

plt.scatter(s.pieces, s.space,label="space")
plt.scatter(s.pieces, s.state,label="current")
plt.scatter(s.pieces, s.random,label="random")


# ax.scatter(s.pieces, s.state,c="blue")

# ax.bar(s.pieces, s.state)
plt.grid()
plt.semilogy()
plt.legend()
# fig.tight_layout()
# ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
#        ylim=(0, 8), yticks=np.arange(1, 8))
plt.show()
```
